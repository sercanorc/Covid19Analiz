## 1-) En çok vaka görülen 5 ülke

```python
dff=final_df.sort_values(by = ['Confirmed'], ascending = [False] ).head(5)
dff.reset_index(inplace=True)
sns.set(style="whitegrid")
plt.rcParams['figure.figsize']='15,6'
plt.title("Onaylanmis vaka sayıları",fontsize=14)
sns.barplot(x='Country/Region', y='Confirmed', data=dff)
```

## 2-) Hasta ölen ve kurtulan insan sayıları

```python
df_sub=clean_df[clean_df['Date']=='2020-11-11'].groupby(['Date','Country/Region']).sum()
df_sub.reset_index(inplace=True)
plt.figure(figsize=(10,6))
df_sub[['Confirmed','Deaths','Recovered']].agg('sum').plot(kind='bar')
```

## 3-) Amerika ve avrupa vaka ölüm durumu

```python
usa=clean_df[clean_df.WHO_Region == "Americas"]
europe = clean_df[clean_df.WHO_Region =="Europe"]
plt.scatter(usa.Confirmed , usa.Deaths , color="red",label="Americas" )
plt.scatter(europe.Confirmed , europe.Deaths , color="green",label="Europe" )
plt.legend()
plt.xlabel("Vaka")
plt.ylabel("Ölüm")
plt.title("Vaka Ölüm")
plt.show()
```

## 4-) 11 Kasım itibariyle Amerika da eyaletlere göre ölüm sayıları

```python
# group by state
usa_df_temp=usa_df[['Province_State','Date','Deaths']]
usa_df_temp=usa_df_temp[usa_df_temp['Date']=='11/11/20'].groupby(['Date','Province_State']).sum()
dff=usa_df_temp.sort_values(by = ['Deaths'], ascending = [False] ).head(10)
dff.reset_index(inplace=True)
sns.set(style="whitegrid")
plt.rcParams['figure.figsize']='15,6'
plt.title("Ölüm sayıları",fontsize=14)
sns.barplot(x='Province_State', y='Deaths', data=dff)
```

## 6-) Türkiye'nin Günlük Ölüm Sayısı vaka oranı

```python
df_sub=clean_df[['Country/Region','Date','Confirmed','Deaths']]
df_sub=df_sub[df_sub['Country/Region']=='Turkey'].groupby(['Date','Country/Region']).sum()
df_sub.reset_index(inplace=True)
ölümOranı=pd.DataFrame()
ölümOranı['Tarih']=df_sub['Date']
ölümOranı['ölümOranı%']=df_sub['Deaths']/df_sub['Confirmed']*100
plt.figure(figsize=(15,6))
plt.title("Türkiye'nin Günlük Ölüm Oranı Grafiği",fontsize=15,color="black")
sns.lineplot(x="Tarih",y="ölümOranı%",data=ölümOranı,color="red")
```

## 8-)Ülkelere göre Covid 19 yayılımı 

```python
df_sub=final_df[['Country/Region','Confirmed']].groupby(['Country/Region']).sum()
df_sub.reset_index(inplace=True)
count=df_sub.groupby(['Country/Region']).agg('count')
count.reset_index(inplace=True)
max_df=df_sub.groupby(['Country/Region']).agg('max')
max_df.reset_index(inplace=True)
dff=pd.DataFrame()
dff['Country/Region']=max_df['Country/Region']
dff['Yayılım']=max_df['Confirmed']/count['Confirmed']
dff=dff.sort_values(by = ['Yayılım'], ascending = [False]).head(10)
plt.rcParams['figure.figsize']='15,6'
plt.title("Yayılma Oranı",fontsize=12)
sns.barplot(y='Country/Region',x='Yayılım', data=dff,)
```

## 13-) Amerikada eyaletlerin toplam ölüm sayıları sıralaması

```python
query = "SELECT Province_State, SUM(Deaths) as TotalDeaths FROM usa_county_wise GROUP BY Province_State ORDER BY TotalDeaths DESC LIMIT 10"
df = pd.read_sql_query(query,connection)
df
```

```python
plt.title("Eyaletler",fontsize=12)
sns.barplot(x='Province_State',y='TotalDeaths', data=df,palette="Blues_d")
```

## 14-) New York eyaletinde şehirlere göre vaka sayıları

```python
query = "SELECT City, SUM(Confirmed) as TotalConfirmed FROM usa_county_wise where Province_State='New York' GROUP BY City ORDER BY TotalConfirmed DESC LIMIT 15"
df = pd.read_sql_query(query,connection)
df
```

```python
plt.title("NewYork",fontsize=12)
sns.barplot(x='City',y='TotalConfirmed', data=df)
```
